```python
import pandas as pd
import csv
import tkinter as tk
from tkinter import *
```

```python
def Update_Price():
    file = open("stockprice.csv","a")
    date = ent_Date.get()
    ent_Date.delete(0,END)
    ticker = ent_Ticker.get()
    ent_Ticker.delete(0,END)
    price = ent_Price.get()
    ent_Price.delete(0,END)
    newrecord = str(date + "," + ticker + "," + price + "\n")
    file.write(newrecord)
    file.close()
    
# Tạo cửa sổ cập nhật giá chứng khoán
win_Price = Tk()
win_Price.title("Cập nhật giá CK")
win_Price.geometry("300x150")

lb_Date = Label(win_Price, text = "Ngày ")
lb_Date.place(x=30, y=10)

lb_Ticker = Label(win_Price, text = "Mã CK ")
lb_Ticker.place(x=30, y=40)

lb_Price = Label (win_Price, text = "Giá: ")
lb_Price.place(x=30,y=70)

ent_Date = Entry(win_Price, width = 15)
ent_Date.place (x=100, y = 10)
#ent_Date.focus()

ent_Ticker = Entry(win_Price, width = 15)
ent_Ticker.place (x=100, y=40)
ent_Ticker.focus()

ent_Price = Entry(win_Price, width = 15)
ent_Price.place(x= 100, y=70)

but_Update = Button(win_Price, text = "Update", command = Update_Price)
but_Update.place(x=100, y=100)




win_Price.mainloop()
```
